#### 4. Product Inventory Project – создайте приложение, которое управляет инвентаризацией продуктов. Создайте класс продукта, у которого есть цена, идентификатор и количество. Затем создайте класс запасов, который отслеживает различные продукты и может суммировать стоимость запасов.

```python
import random
import json


class Product:
    def __init__(self, title, category, price):
        self.title = title
        self.category = category
        self.price = price
        self.id = random.random()

    def __repr__(self):
        return json.dumps({
            "id": self.id,
            "title": self.title,
            "category": self.category,
            "price": self.price,
        })


class Storage:
    products = []

    def __init__(self):
        pass

    def add(self, *product_info):
        for product, count in product_info:
            self.products += [product for _ in range(count)]

    def delete(self, product):
        if product in self.products:
            self.products.remove(product)
            print(f'{product.title} deleted.')
        else:
            print(f'There is no {product.title} in storage')

    def print_raw_storage(self):
        print(self.products)

    def print_all_products(self):
        info = self.log_products_sorted_by_names(self.products)
        total_storage_price = 0
        for k in info.keys():
            total_storage_price += info[k]["total_price"]

        print(f'\nTotal storage price = ${total_storage_price}')

    def print_products_by_title(self, product_name):
        products = [
            product for product in self.products if product.title == product_name]
        qty = len(products)
        total_price = sum([pr.price for pr in products])
        self.log_product(product_name, qty, products[0].category, total_price)

    def print_products_by_category(self, category):
        products = [p for p in self.products if p.category == category]
        self.log_products_sorted_by_names(products)

    def log_products_sorted_by_names(self, products):
        result = {}
        for product in products:
            if product.title in result.keys():
                result[product.title]["qty"] += 1
                result[product.title]["total_price"] += product.price
            else:
                result[product.title] = {
                    "qty": 1,
                    "total_price": product.price,
                    "category": product.category
                }
        for k in result.keys():
            self.log_product(k, result[k]["qty"],
                             result[k]["category"], result[k]["total_price"])
        return result

    def log_product(self, title, qty, category, total_price):
        print(f'{title}: ')
        print(f'\tQuantity: {qty}')
        print(f'\tCategory: {category}')
        print(f'\tTotal price: ${total_price}')


storage = Storage()

milka = Product("Milka", "chocolate", 1.49)
snickers = Product("Snickers", "chocolate", 0.99)
heinz = Product("Heinz", "sauce", 4.49)
calve = Product("Calve", "sauce", 2.49)

storage.add((milka, 2), (heinz, 13))

storage.print_all_products()
storage.delete(milka)
storage.delete(milka)
storage.delete(milka)
storage.print_all_products()
```
