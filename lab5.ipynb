#### 1. Заполните методы класса Line, чтобы они принимали координаты в виде пары кортежей и возвращали наклон и расстояние линии.

```python
class Line:
    def __init__(self, coor1, coor2):
        self.coor1 = coor1
        self.coor2 = coor2

    def distance(self):
        x1, y1 = self.coor1
        x2, y2 = self.coor2
        return ((x2-x1)**2 + (y2-y1)**2)**1/2

    def slope(self):
        x1, y1 = self.coor1
        x2, y2 = self.coor2
        return (y2 - y1)/(x2 - x1)


line = Line((2, 1), (5, 3))
print(line.slope())
print(line.distance())
```

#### 2. Заполните класс:

```python
import math


class Cylinder:
    def __init__(self, height=1, radius=1):
        self.height = height
        self.radius = radius

    def volume(self):
        return math.pi * self.radius ** 2 * self.height

    def surface_area(self):
        return 2 * self.radius * self.height + 2 * math.pi * self.radius ** 2


cylinder = Cylinder(15, 10)
print(cylinder.volume())
print(cylinder.surface_area())
```

#### 3. Для этой задачи создайте класс банковского счета с двумя атрибутами: владелец (owner) и остаток средств (balance). И также два метода: депозит (deposit) и снять средства (withdraw) В качестве дополнительного требования вывод средств не может превышать доступный баланс. Создайте экземпляр своего класса, сделайте несколько депозитов и снятий средств и проверьте, не может ли счет быть перерасходованным.

```python


class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
        self.greet()

    def deposit(self, amount):
        self.balance += amount
        self.log_transaction("deposit", amount)

    def withdraw(self, amount):
        if (self.balance < amount):
            return self.log_error(amount)
        self.balance -= amount
        self.log_transaction("withdraw", amount)

    def greet(self):
        print(f'Hi, {self.owner}! Your balance: ${self.balance}\n')

    def log_transaction(self, action, amount):
        operation = 'withdrawn' if action == "withdraw" else 'deposited'
        print(f'${amount} was {operation}. New balance: {self.balance}')

    def log_error(self, amount):
        print(
            f'\nSorry, you have not enough money. Balance: ${self.balance}. You tried to withdraw ${amount}')


bank_account = BankAccount("Artem", 100)
bank_account.deposit(500)
bank_account.withdraw(200)
bank_account.withdraw(2000)
```

#### 4. Product Inventory Project – создайте приложение, которое управляет инвентаризацией продуктов. Создайте класс продукта, у которого есть цена, идентификатор и количество. Затем создайте класс запасов, который отслеживает различные продукты и может суммировать стоимость запасов.

```python
import random
import json


class Product:
    def __init__(self, title, category, price):
        self.title = title
        self.category = category
        self.price = price
        self.id = random.random()

    def __repr__(self):
        return json.dumps({
            "id": self.id,
            "title": self.title,
            "category": self.category,
            "price": self.price,
        })


class Storage:
    products = []

    def __init__(self):
        pass

    def add(self, *product_info):
        for product, count in product_info:
            self.products += [product for _ in range(count)]

    def delete(self, product):
        if product in self.products:
            self.products.remove(product)
            print(f'{product.title} deleted.')
        else:
            print(f'There is no {product.title} in storage')

    def print_raw_storage(self):
        print(self.products)

    def print_all_products(self):
        info = self.log_products_sorted_by_names(self.products)
        total_storage_price = 0
        for k in info.keys():
            total_storage_price += info[k]["total_price"]

        print(f'\nTotal storage price = ${total_storage_price}')

    def print_products_by_title(self, product_name):
        products = [
            product for product in self.products if product.title == product_name]
        qty = len(products)
        total_price = sum([pr.price for pr in products])
        self.log_product(product_name, qty, products[0].category, total_price)

    def print_products_by_category(self, category):
        products = [p for p in self.products if p.category == category]
        self.log_products_sorted_by_names(products)

    def log_products_sorted_by_names(self, products):
        result = {}
        for product in products:
            if product.title in result.keys():
                result[product.title]["qty"] += 1
                result[product.title]["total_price"] += product.price
            else:
                result[product.title] = {
                    "qty": 1,
                    "total_price": product.price,
                    "category": product.category
                }
        for k in result.keys():
            self.log_product(k, result[k]["qty"],
                             result[k]["category"], result[k]["total_price"])
        return result

    def log_product(self, title, qty, category, total_price):
        print(f'{title}: ')
        print(f'\tQuantity: {qty}')
        print(f'\tCategory: {category}')
        print(f'\tTotal price: ${total_price}')


storage = Storage()

milka = Product("Milka", "chocolate", 1.49)
snickers = Product("Snickers", "chocolate", 0.99)
heinz = Product("Heinz", "sauce", 4.49)
calve = Product("Calve", "sauce", 2.49)

storage.add((milka, 2), (heinz, 13))

storage.print_all_products()
storage.delete(milka)
storage.delete(milka)
storage.delete(milka)
storage.print_all_products()
```

#### 5. Система бронирования авиакомпаний/отелей. Создайте систему бронирования, которая бронирует места в самолетах или номера в отелях. Он взимает различные тарифы за определенные участки самолета или отеля. Например, первый класс будет стоить больше, чем эконом. В гостиничных номерах есть пентхаусы, которые стоят дороже. Следите за тем, когда комнаты будут доступны и могут быть запланированы.

```python
class Hotel:
    def __init__(self, title: str):
        self.title = title

    def __repr__(self):
        return f'Hotel \'{self.title}\''


class Airline:
    def __init__(self, title: str):
        self.title = title

    def __repr__(self):
        return f'Airline \'{self.title}\''


class Supervisor:
    def __init__(self, hotels: list[Hotel] = [], airlines: list[Airline] = []):
        self._hotels = hotels
        self._airlines = airlines

    def add_hotel(self, hotel: Hotel):
        if hotel in self._hotels:
            return print(f'Oops \'{hotel.title}\' already on the list.')
        self._hotels += [hotel]

    def add_airline(self, airline: Airline):
        if airline in self._airlines:
            return print(f'Oops \'{airline.title}\' already on the list.')
        self._airlines += [airline]

    def remove_hotel(self, hotel_title: str):
        hotel_to_remove = next(
            (hotel for hotel in self._hotels if hotel.title == hotel_title), None)

        if hotel_to_remove is None:
            return print(f'There\'s no hotel {hotel_title} on the list.')

        self._hotels.remove(hotel_to_remove)
        print(f'Removed hotel {hotel_title}')

    def remove_airline(self, airline_title: str):
        airline_to_remove = next(
            (airline for airline in self._airlines if airline.title == airline_title), None)

        if airline_to_remove is None:
            return print(f'There\'s no airline {airline_title} on the list.')

        self._airlines.remove(airline_to_remove)
        print(f'Removed airline {airline_title}')

    def get_hotels(self):
        return self._hotels

    def get_airlines(self):
        return self._airlines


mariott = Hotel("Mariott")
s7 = Airline("S7")

sv = Supervisor([mariott], [s7])
print(sv.get_hotels())
print(sv.get_airlines())
sv.remove_hotel(mariott.title)
sv.remove_hotel(mariott.title)
print(sv.get_hotels())
sv.remove_airline(s7.title)
sv.remove_airline(s7.title)
print(sv.get_airlines())
```
